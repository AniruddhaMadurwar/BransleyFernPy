```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
def transformFunc1(x,y):
    return (0.,0.16*y)

def transformFunc2(x,y):
    return (0.85*x + 0.04*y,-0.04*x + 0.85*y + 1.60)

def transformFunc3(x,y):
    return (0.20*x - 0.26*y,0.23*x + 0.22*y + 1.60)

def transformFunc4(x,y):
    return (-0.15*x + 0.28*y,0.26*x + 0.24*y + 0.44)

listOfFunctions = [transformFunc1,transformFunc2,transformFunc3,transformFunc4]
```

```python
datapoints = 100000
x,y = 0,0
x_list = []
y_list = []

for i in range(datapoints):
    functionToCall = np.random.choice(listOfFunctions, p =[0.01,0.85,0.07,0.07])
    x,y = functionToCall(x,y)
    x_list.append(x)
    y_list.append(y)
```

```python
plt.scatter(x_list,y_list, color = 'green')
```

```python
plt.scatter(x_list,y_list,s=0.1, color = 'green')
```

```python
width, height = 300, 300
fern_better = np.zeros((width, height))
x,y = 0, 0 
for i in range(datapoints):
    functionToCall = np.random.choice(listOfFunctions, p =[0.01, 0.85, 0.07, 0.07])
    x, y = functionToCall(x, y)
    ix, iy = int(width / 2 + x * width / 10), int(y* height / 12)
    fern_better[iy, ix] = 1
plt.imshow(fern_better[::-1, :], cmap= cm.Greens)
plt.show()
plt.close()
```

```python
def transformFunc1(x,y):
    return (0.,0.25*y - 0.4)

def transformFunc2(x,y):
    return (0.95*x + 0.005*y -0.002,-0.005*x + 0.93*y + 0.5)

def transformFunc3(x,y):
    return (0.035*x - 0.2*y -0.09, 0.16 * x + 0.04 * y + 0.02)

def transformFunc4(x,y):
    return (-0.04 * x + 0.2 * y +0.083, 0.16 * x + 0.04 * y + 0.12)

listOfFunctions = [transformFunc1,transformFunc2,transformFunc3,transformFunc4]
```

```python
datapoints = 100000
x,y = 0,0
x_list = []
y_list = []

for i in range(datapoints):
    functionToCall = np.random.choice(listOfFunctions, p =[0.02,0.84,0.07,0.07])
    x,y = functionToCall(x,y)
    x_list.append(x)
    y_list.append(y)
    
plt.scatter(x_list,y_list, s=0.2, color = 'green')
```

```python
width, height = 300, 300
fern_mutated = np.zeros((width, height))
x,y = 0, 0 
for i in range(datapoints):
    functionToCall = np.random.choice(listOfFunctions, p =[0.02, 0.84, 0.07, 0.07])
    x, y = functionToCall(x, y)
    ix, iy = int(width / 2 + x * width / 10), int(height/10 + y* height / 12)
    fern_mutated[iy, ix] = 1
plt.imshow(fern_mutated[::-1 , :], cmap= cm.Greens)
plt.show()
```
